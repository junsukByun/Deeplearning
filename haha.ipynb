```python
def Prob_list_graph(self, bin_num=20, hist=True, norm_hist=False, custom_fname=None):
    NonKiller_prob = self.df_type.pred_1

    self.NK_prob = NonKiller_prob[np.where(self.true == 1)[0]]
    self.K_prob = NonKiller_prob[np.where(self.true == 0)[0]]

    if self.fatal_killer_model_yn == True:
        sns.distplot(self.NK_prob , hist=hist, color="skyblue", label="Nonfatal Killer + NonKiller(OK)", bins=bin_num, norm_hist=norm_hist, axlabel="prob_fatal")
        sns.distplot(self.K_prob , hist=hist, color="red", label="Fatal Killer", bins=bin_num, norm_hist=norm_hist, axlabel="prob_fatal")
    else:
        sns.distplot(self.NK_prob , hist=hist, color="skyblue", label="NonKiller(OK)", bins=bin_num, norm_hist=norm_hist, axlabel="prob_OK")
        sns.distplot(self.K_prob , hist=hist, color="red", label="Killer(NG)", bins=bin_num, norm_hist=norm_hist, axlabel="prob_OK")
    plt.title(self.model_name, y=1.02)
    plt.legend()
    if custom_fname:
        plt.savefig(custom_fname+'.png')
    else:
        plt.savefig(self.model_name+'.png')
    plt.show()
```
